```python
# leer archivos 'data/IPH_robos_2021_2024_tecmty.xlsx' y 'data/IPH_robos_ene-ago-2025_tecmty.xlsx'
import pandas as pd

df_2021_2024 = pd.read_excel('data/IPH_robos_2021_2024_tecmty.xlsx')
df_ene_ago_2025 = pd.read_excel('data/IPH_robos_ene-ago-2025_tecmty.xlsx')
```

```python
# cambiar el tipo de dato de la columna 'Fecha' a datetime, y que los datos "<Null>" se conviertan en NaT
df_2021_2024['FECHA'] = pd.to_datetime(df_2021_2024['FECHA'], format='%d/%m/%Y', errors='coerce')
df_ene_ago_2025['FECHA'] = pd.to_datetime(df_ene_ago_2025['FECHA'], format='%d/%m/%Y', errors='coerce')
```

```python
df_2021_2024.head()
```

```python
# cambiar nombre a columna 'VOLENCIA' por 'VIOLENCIA'
df_ene_ago_2025.rename(columns={'VOLENCIA': 'VIOLENCIA'}, inplace=True)
df_ene_ago_2025.head()
```

```python
# hacer concat de los dos dataframes, que el df_ene_ago_2025 se agregue al principio del df_2021_2024
df_robos = pd.concat([df_ene_ago_2025, df_2021_2024], ignore_index=True)
df_robos.head()
```

```python
# ver estadísticas descriptivas de df_robos, en específico la columna 'FECHA'
df_robos['FECHA'].describe(include='all')
```

```python
df_robos.shape
```

```python
# ver renglones donde FECHA es NaT
df_robos[df_robos['FECHA'].isna()]
```

```python
# ver si hay duplicados en las columna de FOLIO
print('La cantidad de datos duplicados en la columna FOLIO es:', df_robos['FOLIO'].duplicated().sum())
```

```python
# Crear columna MES a partir de la columna FECHA
df_robos['MES'] = df_robos['FECHA'].dt.month
df_robos.head()
```

```python
#creamos columna estación del año, tomando en cuenta los datos nulos
def estacion_del_año(mes):
    if pd.isnull(mes):
        return None
    elif mes in [12, 1, 2]:
        return 'Invierno'
    elif mes in [3, 4, 5]:
        return 'Primavera'
    elif mes in [6, 7, 8]:
        return 'Verano'
    elif mes in [9, 10, 11]:
        return 'Otoño'

df_robos['ESTACION'] = df_robos['FECHA'].dt.month.apply(estacion_del_año)
df_robos.head()
```

```python
# ver renglones donde ESTACION es NaN
df_robos[df_robos['ESTACION'].isna()]
```

```python
# crear columna franja horaria, teniendo en cuenta que hay horas con valor de 99 (nulo)
def franja_horaria(hora):
    if hora == 99:
        return None
    elif 0 <= hora < 6:
        return 'Madrugada'
    elif 6 <= hora < 12:
        return 'Mañana'
    elif 12 <= hora < 18:
        return 'Tarde'
    elif 18 <= hora < 24:
        return 'Noche'
    else:
        return 'Desconocido'

df_robos['FRANJA HORARIA'] = df_robos['HORA'].apply(franja_horaria)
df_robos.head()
```

```python
# crear columna de VIOLENCIA_BIN donde si es con violencia es 1 y sin violencia es 0
df_robos['VIOLENCIA_BIN'] = df_robos['VIOLENCIA'].apply(lambda x: 1 if x == 'SI' else 0)
df_robos.head()
```

```python
# crear columna ROBO A NEGOCIO donde si el TIPO es 'ROBO A NEGOCIO' es 1 y si no es 0
df_robos['ROBO A NEGOCIO'] = df_robos['TIPO'].apply(lambda x: 1 if x == 'ROBO A NEGOCIO' else 0)
df_robos.head()
```

```python
# crear columna ROBO A CASA HABITACION donde si el TIPO es 'ROBO A CASA HABITACION' es 1 y si no es 0
df_robos['ROBO A CASA HABITACION'] = df_robos['TIPO'].apply(lambda x: 1 if x == 'ROBO A CASA HABITACION' else 0)
df_robos.head()
```

```python
# crear columna ROBO DE VEHICULO donde si el TIPO es 'ROBO DE VEHICULO' es 1 y si no es 0
df_robos['ROBO DE VEHICULO'] = df_robos['TIPO'].apply(lambda x: 1 if x == 'ROBO DE VEHICULO' else 0)
df_robos.head()
```

```python
# GRAFICAR cantidad de ROBO A NEGOCIO por franja horaria y estación del año
df_robos[df_robos['ROBO A NEGOCIO'] == 1].groupby(['FRANJA HORARIA', 'ESTACION']).size().unstack(fill_value=0).plot(kind='bar', stacked=True)
```

```python
# leer archivo data/IPH_robos_2021-2025_tecmty (1).xlsx
df_robos_final = pd.read_excel('data/IPH_robos_2021-2025_tecmty (1).xlsx')
df_robos_final.head()
```

```python
# crear columna MES a partir de la columna FECHA
df_robos_final['MES'] = df_robos_final['FECHA'].dt.month
df_robos_final.head()
```

```python
# ver si haay datos nulos en la columna MES
df_robos_final[df_robos_final['MES'].isna()]
```

```python
# crear columna estacion del año a partir de la columna MES
def estacion_del_año_final(mes):
    if pd.isnull(mes):
        return None
    elif mes in [12, 1, 2]:
        return 'Invierno'
    elif mes in [3, 4, 5]:
        return 'Primavera'
    elif mes in [6, 7, 8]:
        return 'Verano'
    elif mes in [9, 10, 11]:
        return 'Otoño'
df_robos_final['ESTACION'] = df_robos_final['MES'].apply(estacion_del_año_final)
df_robos_final.head()
```

```python
# graficar cantidad de robos por estación del año por cuadrante
df_robos_final.groupby(['ESTACION', 'CUADRANTE']).size().unstack(fill_value=0).plot(kind='bar', stacked=True)
```

```python
# ver cantidad de robos por mes, que el titulo de la gráfica sea "Cantidad de robos por mes"
df_robos_final.groupby('MES').size().plot(kind='bar', title='Cantidad de robos por mes')
```

```python
# ver cantidad de robos a negocios por mes, que el titulo de la gráfica sea "Cantidad de robos a negocios por mes"
df_robos_final[df_robos_final['TIPO'] == 'ROBO A NEGOCIO'].groupby('MES').size().plot(kind='bar', title='Cantidad de robos a negocios por mes')
```

La primera gráfica presenta el Robo a Negocio.
Podemos observar que hay picos claros en marzo y otro en agosto, y una caída relativa en septiembre.
Esta gráfica sugiere estacionalidad, es decir, aumentos en primavera y finales de verano. Podría relacionarse con mayor movilidad o eventos específicos en esos meses.
Se recomendaría focalizar prevención en los meses pico y revisar campañas de vigilancia.

```python
# Ver cantidad de robos a casa habitación por mes, que el título sea "Cantidad de robos a casa habitación por mes"
df_robos_final[df_robos_final['TIPO'] == 'ROBO A CASA HABITACION'].groupby('MES').size().plot(kind='bar', title='Cantidad de robos a casa habitación por mes')
```

La segunda gráfica muestra los Robos a Casa Habitación por mes. Aquí el patrón es similar: pico fuerte en marzo/abril y valores bajos en enero.”
Comparada con la primera, el mínimo es más marcado en enero y el pico en marzo/abril es igual o mayor.
Se recomendaría en este caso investigar diferencias locales (zonas, horarios) para adaptar medidas preventivas.

```python
# ver cantidad de robos de vehículo por mes, que el titulo sea "Cantidad de robos de vehículo por mes"
df_robos_final[df_robos_final['TIPO'] == 'ROBO DE VEHICULO'].groupby('MES').size().plot(kind='bar', title='Cantidad de robos de vehículo por mes')
```

La tercera gráfica corresponde a Robo de Vehículos y tiene volúmenes totales más altos. Se podría decir que no hay mucho cambio de mes a mes en este tipo de robo. El máximo está en agosto (al igual que en el robo a negocio) y hay una caída pronunciada en septiembre (al igual que en robo a negocio).
